```python
import numpy as np
import pandas as pd
import scipy.io as sc
import matplotlib.pyplot as plt
```

```python
data = sc.loadmat("data.mat")
X = data['X']
y = data['yval']
print(' X\t\t\t   Xval\t\t\t     y')
for x, y, z in zip(X, data['Xval'], y):
    print(x, y, z)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.grid(True)
```

```python
meanX1 = np.mean(X[:,0])
meanX2 = np.mean(X[:,1])
print('mean for X1: %f\nmean for X2: %f' % (meanX1, meanX2))

varX1 = np.mean((X[:,0]-meanX1)**2)
varX2 = np.mean((X[:,1]-meanX2)**2)
print('variance for X1: %f\nvariance for X2: %f' % (varX1, varX2))
```

```python
def gaussian(m, var, x):
    return 1/(np.sqrt(2*np.pi*var)*np.exp((x-m)**2/(2*var)))

y_pred = data['yval']
eps = 0.0025
pX1 = gaussian(meanX1, varX1, X[:,0])
pX2 = gaussian(meanX2, varX2, X[:,1])
for i in range(len(X)):
    if pX1[i]*pX2[i]<=eps:
        plt.scatter(X[i,0], X[i,1] ,c='r')
        y_pred[i] = 1
        print((X[i,0], X[i,1]),':', (pX1[i]*pX2[i]))
    else:
        plt.scatter(X[i,0], X[i,1], c='b', alpha=0.4)
        y_pred[i] = 0
```

```python
def f1_score(act, pred):

    tp = len([[x,y] for x,y in zip(act, pred) if x==y and y==1])
    fp = len([[x,y] for x,y in zip(act, pred) if x!=y and x==0])
    fn = len([[x,y] for x,y in zip(act, pred) if x!=y and x==1])
    print([[x,y] for x,y in zip(act, pred) if x==y and y==1])
    print([[x,y] for x,y in zip(act, pred) if x!=y and y==1])
    print([[x,y] for x,y in zip(act, pred) if x!=y and x==1])

    if tp>0:
        p=float(tp)/(tp+fp)
        r=float(tp)/(tp+fn)

        return 200*((p*r)/(p+r))
    else:
        return 0
    
f1_score(data['yval'], y_pred)
```
